## 1.1 SELECT

**Para que serve?** Serve para selecionar as colunas de tableas



---



### **Exemplos**

**(Exemplo 1)** Seleção de uma coluna de uma tabela
* Liste os e-mails dos clientes da tabela sales.customers


**Query:**
```sql
select email
from sales.customers
```



image.png

**(Exemplo 2)** Seleção de mais de uma coluna de uma tabela
* Liste os emails e nomes dos clientes da tabela sales.customers

**Query:**

```sql
select email, first_name, last_name
from sales.customers
```



image.png

**(Exemplo 3)** Seleção de todas as colunas de uma tabela
* Liste todos as informações dos clientes da tabela sales.customers

**Quuery:**

```sql
select *
from sales.customers

```



image.png

### **Resumo**



1.   Comando usado para selecionar colunas de tabelas
2.   Quando selecionar mais de uma coluna, elas devem ser separadas por vírgula sem conter vírgula antes do comando FROM
3.   Pode-se utilizar o asterisco (*) para selecionar todas as colunas da tabela



## 1.2 **DISTINCT**

**Para que serve?**
* Serve para remover linhas duplicadas e mostrar apenas linhas distintas
* Muito usado na etapa de exploração de dados.

**Sintaxe**

```sql
select distinct coluna_1, coluna_2, coluna_3
from schema_1.tabela_1
```



### **Exemplos**

**(Exemplo 1)** Seleção de uma coluna sem DISTINCT
* Liste as marcas de carro que constam na tabela products

**Query:**

```sql
select brand
from sales.products
```



image.png

> Note que tivemos 333 linhas retornadas.

**(Exemplo 2)** Seleção de uma coluna com DISTINCT
* Liste as marcas de carro distintas que constam na tabela products

Agora a mesma consulta utilizando o **DISTINCT**

**Query:**

```sql
select distinct brand
from sales.products
```



image.png

> Note que agora tivemos apenas 40 linhas com a consulta.

**(Exemplo 3)** Seleção de mais de uma coluna com DISTINCT
* Liste as marcas e anos de modelo distintos que constam na tabela products

Agora faremos uma consulta "complexa" com mais de 1 coluna para ver o resultado.

**Query:**

```sql
select distinct brand, model_year
from sales.products
```



image.png

> Note que agora tivemos 184 linhas com a consulta, **porque?** porque o distinct levou em consideração o ano do modelo, pois podemos notar que existe "Ford 2014 e Ford 2010" por exemplo.

### **Resumo**

1.  Comando usado para remover linhas duplicadas e mostrar apenas linhas distintas
2. Muito utilizado na etapa de exploração dos dados
3. Caso mais de uma coluna seja selecionada, o comando SELECT DISTINCT irá retornar todas as combinações distintas.

## 1.3 WHERE

**Para que serve?**
* Serve para filtrar linhas de arcodo com uma condição.

**Sintaxa**

```sql
select coluna_1, coluna_2, coluna_3
from schema_1.tabela_1
where condição_x=true
```



### **Exemplos**

**(Exemplo 1)** Filtro com condição única
* Liste os emails dos clientes da nossa base que moram no estado de Santa Catarina

**Query:**

```sql
select email, state
from sales.customers
where state = 'SC'
```



image.png

> Notamos que apenas os emails dos usarias de Santa Satarina apareceu na tabela pos a query.

**(Exemplo 2)** Filtro com mais de uma condição
* Liste os emails dos clientes da nossa base que moram no estado de Santa Catarina ou Mato Grosso do Sul

**Query:**

```sql
select email, state
from sales.customers
where state = 'SC' or state = 'MS'
```



image.png

> Olhe que legal, agora na tabela tambem aparece os usuarios de Mato Grosso do Sul.

 **(Exemplo 3)** Filtro de condição com data
* Liste os emails dos clientes da nossa base que moram no estado de Santa Catarina ou Mato Grosso do Sul e que tem mais de 30 anos

image.png

> Analisando o banco de dados percebemos que não existe o idade do usuario, apenas a data de nascimento, iremos verificar o formato da data com uma query.

Query:

```sql
select distinct birth_date
from sales.customers
```



image.png

agora que temos o formato podemos processeguir com a query.

**Query:**

```sql
select email, state, birth_date
from sales.customers
where (state= 'SC' or state = 'MS') and birth_date < '1991-12-28'
```

```sql
select email, state, birth_date
from sales.customers
where (state= 'SC' or state = 'MS') and birth_date < '19911228'
```



Ambas as querys irá retornar o mesmo resultado, podemos omitir ( - )

image.png

> Como esperado, o resultado foi os emails dos usuarios de Santa Catarina e Mato Grosso do Sul, nascidos antes de `'1991-12-28'`

### **Resumo**



1.   Comando utilizado para filtrar linhas de acordo com uma condição
2.   No PostgreSQL são utilizadas aspas simples para delimitar strings
1.   string = sequência de caracteres = texto
1.   Pode-se combinar mais de uma condição utilizando os operadores lógicos
2.   No PostgreSQL as datas são escritas no formato 'YYYY-MM-DD' ou 'YYYYMMDD'






## **1.4 ORDER BY**

Para que serve?
* Serve para ordenar a seleção de acordo com uma regra definida pelo usuário.

**Sintaxe**

```sql
select coluna_1, coluna_2, coluna_3
from schema_1.tabela_1
where condição_x=true
order by coluna_1
```



### **EXEMPLOS**

**(Exemplo 1)** Ordenação de valores numéricos
* Liste produtos da tabela products na ordem crescente com base no preço

**Query:**

```sql
select price
from sales.products
order by price
```



image.png

> perceba que os preços dos produtos ficaram ordenados de forma crescente, é possivel deixa-los de forma decrescente? sim, vamos ver agora.

**Query:**

```sql
select price
from sales.products
order by price desc
```



image.png

> Olha que legal, agora ficou ordenado da forma decrescente apenas adiconando o `DESC` a query.

**(Exemplo 2**) Ordenação de texto
* Liste os estados distintos da tabela customers na ordem crescente

**Query:**

```sql
select state distinct
from sales.customers
order by state
```



image.png

> Olha que legal, orde by em colunas do tipo "String" ficam em ordem alfabetica, como sera que fica na forma decrescente? veremos agora!

**Query:**

```sql
select state distinct
from sales.customers
order by state desc
```



image.png

> Realmente ficou ordenado de forma inversa, como esperado.

### **Resumo**



1.   Comando utilizado para ordenar a seleção de acordo com uma regra definida
2.   Por padrão o comando ordena na ordem crescente. Para mudar para a ordem decrescente usar o comando DESC
3.  No caso de strings a ordenação será seguirá a ordem alfabetica



## **1.5 LIMIT**

Para que serve?
* Serve para limitar o nº de linhas da consulta.
* Muito utilizado na etapa de exploração dos dados


**Sintaxe**


```sql
select coluna_1, coluna_2, coluna_3
from schema_1.tabela_1
limit N
```



### **EXEMPLOS**

**(Exemplo 1)** Seleção das N primeiras linhas usando LIMIT
* Liste as 5 primeiras linhas da tabela funnel utilizando o LIMIT

**Query:**

```sql
select *
from sales.funnel
limit 5
```



image.png

> Perceba que foi limitado em apenas 5 linhas a consulta.

**(Exemplo 2)** Seleção das N primeiras linhas usando LIMIT e ORDER BY
* Liste os 5 produtos mais caros da tabela products com o comando LIMIT

**Query:**

```sql
select *
from sales.products
order by price desc
limit 5
```



image.png

> Perceba, agrupamos apenas os 5 produtos mais caros da tabela.

### **Resumo**



1.   Comando utilizado para limitar o nº de linhas da consulta.
2.    Muito utilizado na etapa de exploração dos dados
3.    Muito utilizado em conjunto com o comando ORDER BY quando o que importa são os TOP N. Ex: "N pagamentos mais recentes", "N produtos mais caros"

**Query:**

```sql
select city
from sales.customers
order by city
```



image.png

**(Exercício 2)** Selecione o visit_id das 10 compras mais recentes efetuadas (dados da tabela sales.funnel)

**Query:**

```sql
SELECT visit_id, paid_date
FROM sales.funnel
ORDER BY paid_date DESC
LIMIT 10
```



image.png

> Essa consulta na forma `DESC` esta retornando valores nulos, precisamos tratar isso com `WHERE`.

**Query:**

```sql
SELECT visit_id, paid_date
FROM sales.funnel
WHERE paid_date IS NOT NULL
ORDER BY paid_date DESC
LIMIT 10
```



image.png

> Agora sim! conseguimos os 10 pagamentos mais recente no banco de dados.

**(Exercício 3)** Selecione todos os dados dos 10 clientes com maior score nascidos após 01/01/2000 (dados da tabela sales.customers)

**Query:**

```sql
SELECT *
FROM sales.customers
WHERE birth_date < '20000101'
LIMIT 10
```



image.png

> Note que conseguirmos os 10 clientes com maior score nascido antes dos anos 2000.
